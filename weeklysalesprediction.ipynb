```python
from sklearn.preprocessing import OneHotEncoder
def remove_outliers_series(series):

    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    series_cleaned = series[~((series < lower_bound) | (series > upper_bound))]
    return series_cleaned

for i in columns:
    data[i]=remove_outliers_series(data[i])

data=data.dropna()

dates=data["Date"]
data=data.drop("Date",axis=1)
dates=pd.to_datetime(dates, format="%d-%m-%Y")

data=pd.get_dummies(data, columns=["Store"],drop_first=True)
# data["dates"]=dates
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
X=pd.DataFrame(data.drop("Weekly_Sales", axis=1))
y=pd.DataFrame(np.array(data['Weekly_Sales']))
dates=pd.DataFrame(dates)

x_sc= pd.DataFrame(scaler.fit_transform(X))
y_sc= pd.DataFrame(scaler.fit_transform(y))



x_sc["dates"]=dates
X["dates"]=dates



```

```python
data.head(-1)
```

```python
if isinstance(x_sc, pd.DataFrame):
    x_sc = x_sc.select_dtypes(include=[np.number])


x_sc = np.hstack((np.array(x_sc), np.ones((x_sc.shape[0], 1))))
Y_predict = np.array(y_sc)

for lambda_ in [0.1,1]:
    I = np.eye(x_sc.shape[1])
    I[0, 0] = 0
    W_g = np.linalg.inv(x_sc.T @ x_sc + lambda_*I) @ x_sc.T @ Y_predict
    print(f"Lambda: {lambda_}, Weights: {W_g}")

```

```python
from sklearn.model_selection import train_test_split
x_sc,x_sc_test,Y_predict,y_test=train_test_split(x_sc,Y_predict,test_size=0.2)
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import LassoCV


ridge_cv = Ridge()
ridge_cv.fit(x_sc, Y_predict)

lasso_cv = LassoCV(alphas=[0.001], cv=5)
lasso_cv.fit(x_sc, Y_predict)

print("Best Lasso Lambda:", ridge_cv)
print("Lasso Weights:", ridge_cv.coef_)
print("Best Lasso Lambda:", lasso_cv.alpha_)
print("Lasso Weights:", lasso_cv.coef_)
```

```python
from sklearn.metrics import mean_squared_error as mse
# Predict values
y_pred_normal = x_sc @ W_g  # W from Normal Equation
y_pred_ridge= ridge_cv.predict(x_sc)
y_pred_lasso = lasso_cv.predict(x_sc)

# Compute errors
mse_normal = mse(Y_predict, y_pred_normal)
mse_ridge = mse(Y_predict, y_pred_ridge)
mse_lasso=mse(Y_predict, y_pred_lasso)

print(f"Normal Equation MSE:  {mse_normal}")
print(f"Ridge Equation MSE:  {mse_ridge}")
print("Best Lasso Lambda:", lasso_cv.alpha_)
print(f"Lasso MSE:", mse_lasso)
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

y_pred_normalrmse= rmse(y_pred_normal,Y_predict)
y_pred_ridgermse= rmse(y_pred_ridge,Y_predict)


print(Y_predict)
print(y_pred_normalrmse)
print(y_pred_ridgermse)
```

```python
def mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

def gradient_descent(W, Input, output, epoches, learning_rate):
    for i in range(epoches):
        y_pred = Input @ W
        error= rmse(y_pred, output)

        gradient = (Input.T @ (y_pred - output)) / len(output)

        W = W - learning_rate * gradient

        if i % 1000 == 0:
            print(f"Epoch {i}, MSE: {error}")

    return W


w_g=gradient_descent(W_g,x_sc,Y_predict,20000,0.8)
```
